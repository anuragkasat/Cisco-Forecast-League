```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
#from fbprophet import Prophet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error

# Assuming df_filtered is your DataFrame with updated demand data

results = []

for product_name in df_filtered['Product Name'].unique():
    product_data = df_filtered[df_filtered['Product Name'] == product_name]
    historical_demand = pd.to_numeric(product_data['FY23 Q3'], errors='coerce')

    # SARIMA
    model_sarima = SARIMAX(historical_demand, order=(1, 1, 1), seasonal_order=(1, 1, 1, 4)).fit()
    forecast_sarima = model_sarima.get_forecast(steps=1).predicted_mean.iloc[0]
    mae_sarima = mean_absolute_error([historical_demand.iloc[-1]], [forecast_sarima])
    results.append({'Product Name': product_name, 'SARIMA Forecast': forecast_sarima, 'MAE_SARIMA': mae_sarima})

    # Prophet
    df_prophet = pd.DataFrame({'ds': historical_demand.index, 'y': historical_demand.values})
    model_prophet = Prophet(yearly_seasonality=True)
    model_prophet.fit(df_prophet)
    future_prophet = model_prophet.make_future_dataframe(periods=1)
    forecast_prophet = model_prophet.predict(future_prophet).yhat.iloc[-1]
    mae_prophet = mean_absolute_error([historical_demand.iloc[-1]], [forecast_prophet])
    results.append({'Product Name': product_name, 'Prophet Forecast': forecast_prophet, 'MAE_Prophet': mae_prophet})

    # Random Forest
    model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
    model_rf.fit(historical_demand.index.values.reshape(-1, 1), historical_demand.values)
    forecast_rf = model_rf.predict([[historical_demand.index[-1] + 1]])
    mae_rf = mean_absolute_error([historical_demand.iloc[-1]], [forecast_rf])
    results.append({'Product Name': product_name, 'RF Forecast': forecast_rf[0], 'MAE_RF': mae_rf})

    # Gradient Boosting
    model_gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
    model_gb.fit(historical_demand.index.values.reshape(-1, 1), historical_demand.values)
    forecast_gb = model_gb.predict([[historical_demand.index[-1] + 1]])
    mae_gb = mean_absolute_error([historical_demand.iloc[-1]], [forecast_gb])
    results.append({'Product Name': product_name, 'GB Forecast': forecast_gb[0], 'MAE_GB': mae_gb})

    # Neural Network
    model_nn = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
    model_nn.fit(historical_demand.index.values.reshape(-1, 1), historical_demand.values)
    forecast_nn = model_nn.predict([[historical_demand.index[-1] + 1]])
    mae_nn = mean_absolute_error([historical_demand.iloc[-1]], [forecast_nn])
    results.append({'Product Name': product_name, 'NN Forecast': forecast_nn[0], 'MAE_NN': mae_nn})

# Display the results
df_results = pd.DataFrame(results)
print(df_results)
```
